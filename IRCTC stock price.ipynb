```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as pt
df = pd.read_excel(r"C:\Users\Admin\Desktop\sem 5\ML\ML lab\Lab Session1 Data.xlsx",sheet_name="IRCTC Stock Price")
print(df)

pricedata = df['Price']
m = np.mean(pricedata)
v = np.var(pricedata)
print('Mean = ', m)
print('Variance = ', v)

wed_d = pricedata[df['Day'] == 'Wed']
s_mean = np.mean(wed_d)
print('Sample mean = ', s_mean)
a_d = pricedata[df['Month'] == 'Apr']
a_mean = np.mean(a_d)
print('April mean = ', a_mean)

chgd = df['Chg%']
print(chgd)
l = np.where(chgd > 0, False, True)

prob_of_loss = np.mean(l)
print('Probability of loss:', prob_of_loss)
wed_d = df[df['Day'] == 'Wed']
prob_of_profit_wed = np.mean(wed_d['Chg%'] > 0)
print('Probability of profit on Wednesday:', prob_of_profit_wed)
prob_of_wed = np.mean(df['Day'] == 'Wed')

c_prob = prob_of_profit_wed / prob_of_wed
print('Conditional probability of profit which is wednesday:', c_prob)

pt.scatter(df['Day'], df['Chg%'])
pt.xlabel('Day')
pt.ylabel('Chg%')
pt.title('Chg% vs Day')
pt.show()
```
