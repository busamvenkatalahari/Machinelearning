```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
 
# Load the dataset from an Excel file
# Make sure the Excel file is in the same directory as your script or provide the full path to the file
data = pd.read_excel(r"C:\Users\Admin\Desktop\SEM\sem 5\ML\GLCM_extended_dataset_31-01-2023.xlsx")
 
# Extract the input feature (X) and the target variable (y)
GLCM_extended_dataset_X = data[['dissimilarity_0']]  # Replace 'YourFeatureColumn' with the actual column name
GLCM_extended_dataset_y = data['label']  # Replace 'YourTargetColumn' with the actual column name
 
# Use only one feature (adjust column index as needed)
# For example, if you want to use the second column as the feature:
# your_dataset_X = your_dataset_X.iloc[:, 1]
 
# Split the data into training/testing sets
GLCM_extended_dataset_X_train = GLCM_extended_dataset_X[:-20]
GLCM_extended_dataset_X_test = GLCM_extended_dataset_X[-20:]
 
GLCM_extended_dataset_y_train = GLCM_extended_dataset_y[:-20]
GLCM_extended_dataset_y_test = GLCM_extended_dataset_y[-20:]
 
# Create linear regression object
regr = linear_model.LinearRegression()
 
# Train the model using the training sets
regr.fit(GLCM_extended_dataset_X_train, GLCM_extended_dataset_y_train)
 
# Make predictions using the testing set
GLCM_extended_dataset_y_pred = regr.predict(GLCM_extended_dataset_X_test)
 
# The coefficients
print("Coefficients: \n", regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(GLCM_extended_dataset_y_test, GLCM_extended_dataset_y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(GLCM_extended_dataset_y_test, GLCM_extended_dataset_y_pred))
 
# Plot outputs
plt.scatter(GLCM_extended_dataset_X_test, GLCM_extended_dataset_y_test, color="black")
plt.plot(GLCM_extended_dataset_X_test, GLCM_extended_dataset_y_pred, color="blue", linewidth=3)
 
plt.xticks(())
plt.yticks(())
 
plt.show()
```
