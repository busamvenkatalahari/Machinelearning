```python
import pandas as pd
import numpy as np
df = pd.read_excel(r"C:\Users\Admin\Desktop\sem 5\ML\ML lab\Lab Session1 Data.xlsx",sheet_name='Purchase data')
df = df.dropna(axis=1)
print(df)
A = df[['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']].values
C = df[['Payment (Rs)']].values
print("Matrix A")
print(A)
print("Matrix C")
print(C)
Au = np.hstack((A, C))
print("Agumented Matrix of A and C")
print(Au)
d=np.linalg.matrix_rank(Au)
print("Dimensionality of the Vector Space")
print(d)
R_A=np.linalg.matrix_rank(A)
print("Rank of Matrix A")
print(R_A)
Ain=np.linalg.pinv(A)
print("Pseudo inverse of matrix A")
print(Ain)
X=np.matmul(Ain,C)
print("Matrix X")
print(X)
```
